#MOVIMIENTOS MENSUALES POR AÑO

```python
month_HM1 = s1.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM2 = s2.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM3 = s3.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM4 = s4.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM5 = s5.groupby('Año')['Mes'].value_counts().unstack().fillna(0)

frames = [month_HM1, month_HM2, month_HM3, month_HM4, month_HM5]

result = pd.concat(frames)

result
```

Juntamos el número total de los datos por mes

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import pandas as pd

meses = {1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio',
         7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'}

month_HM1 = s1.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM2 = s2.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM3 = s3.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM4 = s4.groupby('Año')['Mes'].value_counts().unstack().fillna(0)
month_HM5 = s5.groupby('Año')['Mes'].value_counts().unstack().fillna(0)

frames = [month_HM1, month_HM2, month_HM3, month_HM4, month_HM5]

result = pd.concat(frames)

result = result.rename(columns=meses)

fig = go.Figure()

for col in result.columns:
    fig.add_trace(go.Scatter(x=result.index, y=result[col], name=col))

fig.update_layout(title='Movimientos Mensuales de los Últimos 5 Años',
                  xaxis_title='Mes',
                  yaxis_title='Cantidad',
                  xaxis_tickmode='array',
                  xaxis_tickvals=result.index.unique())

pyo.iplot(fig)
```

#SUMA DE EGRESOS POR AÑO


```python
d18 = s1.groupby('Mes')['Total'].sum()
d19 = s2.groupby('Mes')['Total'].sum()
d20 = s3.groupby('Mes')['Total'].sum()
d21 = s4.groupby('Mes')['Total'].sum()
d22 = s5.groupby('Mes')['Total'].sum()

print("2018\n\n", d18.apply(lambda x: '${:,.2f}'.format(x)))
print("\n\n2019\n\n", d19.apply(lambda x: '${:,.2f}'.format(x)))
print("\n\n2020\n\n", d20.apply(lambda x: '${:,.2f}'.format(x)))
print("\n\n2021\n\n", d21.apply(lambda x: '${:,.2f}'.format(x)))
print("\n\n2022\n\n", d22.apply(lambda x: '${:,.2f}'.format(x)))
```

```python
import pandas as pd
import plotly.graph_objs as go

d18 = s1.groupby('Mes')['Total'].sum()
d19 = s2.groupby('Mes')['Total'].sum()
d20 = s3.groupby('Mes')['Total'].sum()
d21 = s4.groupby('Mes')['Total'].sum()
d22 = s5.groupby('Mes')['Total'].sum()

df = pd.DataFrame({
    '2018': d18,
    '2019': d19,
    '2020': d20,
    '2021': d21,
    '2022': d22
})

fig = go.Figure()

for column in df.columns:
    fig.add_trace(go.Scatter(x=df.index, y=df[column], name=column))

fig.update_layout(title='Total de egresos por mes y año',
                  xaxis_title='Mes',
                  yaxis_title='Total')

fig.show()
```

```python
print("Egresos mensuales totales de los últimos 5 años:")

sf = [d18, d19, d20, d21, d22]

total = sum(sf)

print(total.apply(lambda x: '${:,.2f}'.format(x)))
```

Al analizar la información de los egresos mensuales de los últimos 5 años, podemos observar que el mes que más gastos tiene es Diciembre.

```python
d18 = s1.groupby('Mes')['Total'].sum()
print("Egresos del 2018: ${:,.2f}".format(d18.sum().round(2)))
print("\n")

d19 = s2.groupby('Mes')['Total'].sum()
print("Egresos del 2019: ${:,.2f}".format(d19.sum().round(2)))
print("\n")

d20 = s3.groupby('Mes')['Total'].sum()
print("Egresos del 2020: ${:,.2f}".format(d20.sum().round(2)))
print("\n")

d21 = s4.groupby('Mes')['Total'].sum()
print("Egresos del 2021: ${:,.2f}".format(d21.sum().round(2)))
print("\n")

d22 = s5.groupby('Mes')['Total'].sum()
print("Egresos del 2022: ${:,.2f}".format(d22.sum().round(2)))
```

```python
import pandas as pd
import plotly.graph_objects as go

totales = pd.Series([
    d18.sum().round(2),
    d19.sum().round(2),
    d20.sum().round(2),
    d21.sum().round(2),
    d22.sum().round(2)
], index=[2018, 2019, 2020, 2021, 2022])

fig = go.Figure()

fig.add_trace(go.Scatter(x=totales.index, y=totales.values, mode='lines+markers'))

fig.update_layout(
    xaxis=dict(type='category', categoryorder='array', categoryarray=[2018, 2019, 2020, 2021, 2022]),
    xaxis_title="Año",
    yaxis_title="Total de Egresos",
    title="Evolución de Egresos por Año"
)

fig.show()
```

Aquí podemos observar que el año con más egresos fue el año de 2019 y el menor el año de 2022

```python
sf = [d18.sum(), d19.sum(), d20.sum(), d21.sum(), d22.sum()]

total_egresos = sum(sf)
total_egresos_str = "{:,.2f}".format(total_egresos)

print("Total de egresos de los últimos 5 años: $", total_egresos_str)
```
